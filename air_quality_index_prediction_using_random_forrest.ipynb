```python
dataframe2 = dataframe.copy()
```

```python
# As there are null values, these values are replaced with the mean
```

```python
dataframe2 = dataframe2.fillna(dataframe2.mean())
```

```python
dataframe2.isnull().sum()
```

```python
dataframe2.info()
```

```python
# to convert the featues which are having the datatype as objects
```

```python
dist = (dataframe2["City"])
distset = set(dist)
dd=list(distset)
dictOfWords = { dd[i] : i for i in range(0, len(dd) ) }
dataframe2["City"] = dataframe2["City"].map(dictOfWords)
```

```python
dist = (dataframe2["AQI_Bucket"])
distset = set(dist)
dd=list(distset)
dictOfWords = { dd[i] : i for i in range(0, len(dd) ) }
dataframe2["AQI_Bucket"] = dataframe2["AQI_Bucket"].map(dictOfWords)
```

```python
dataframe2["AQI_Bucket"]=dataframe2["AQI_Bucket"].fillna(dataframe2['AQI_Bucket'].mean())
```

```python
dataframe2.isnull().sum()
```

```python
dataframe2 = dataframe2.drop('Date', axis=1)
```

```python
dataframe2 = dataframe2.drop('AQI_Bucket',axis=1)
```

Data Pre-Processing and Feature Selection

```python
dataframe2.info()
```

```python
dataframe2.columns
```

```python
features = dataframe2[['City', 'PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3','Benzene', 'Toluene', 'Xylene']]
```

```python
label=dataframe2['AQI']
```

To split the data into Training and Testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(features,label,test_size=0.2,random_state=2)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn import tree
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
regressor = RandomForestRegressor(max_depth=4, random_state=0)
regressor.fit(X_train,Y_train)
```

```python
y_prediction = regressor.predict(X_test)
```

```python
print(y_prediction)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, y_prediction)
```
